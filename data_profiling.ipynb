```python
def check_numeric(data, column):
    assert data[column].dtype in ['int64', 'float64'], f'{column} is not numeric'

numeric_cols = ['Total Interns', 'Employees']

for col in numeric_cols:
    check_numeric(df, col)

print("all the columns are numeric.")
```

```python
def is_outlier(value, q1, q3):
    iqr = q3 - q1
    lower_bound = q1 - 1.5*iqr
    upper_bound = q3 + 1.5*iqr
    return value < lower_bound or value > upper_bound

def column_has_outliers(data, column):
    q1 = data[column].quantile(0.25)
    q3 = data[column].quantile(0.75)
    return data[column].apply(lambda x: is_outlier(x, q1, q3))

assert not column_has_outliers(df, 'Total Interns').any(), 'Total Interns has outliers'
assert not column_has_outliers(df, 'Employees').any(), 'Employees has outliers'
```
